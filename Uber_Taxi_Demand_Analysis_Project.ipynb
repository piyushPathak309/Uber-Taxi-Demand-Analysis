```python
df['Date/Time']=pd.to_datetime(df['Date/Time'])
```

```python
# Adding some useful columns in term of time series
```

```python
df['WeekDay'] = df['Date/Time'].dt.day_name()
```

```python
df['Day'] = df['Date/Time'].dt.day
```

```python
df['Minute'] = df['Date/Time'].dt.minute
df['Month'] = df['Date/Time'].dt.month
df['Year'] = df['Date/Time'].dt.year
df['Hour'] = df['Date/Time'].dt.hour
df['Second'] =  df['Date/Time'].dt.second
```

```python
# Heatmap By Weekday And Hour
```

```python
def HeatMap(col1,col2):
    by_cross = df.groupby([col1,col2]).apply(count_rows).unstack()
    plt.figure(figsize=(16,9))
    return sns.heatmap(by_cross)
```

```python
HeatMap('WeekDay','Hour')
plt.title("Cross Analysis By Hour And WeekDay")
```

```python
#  means On Evening Time Each And Every Day We Have Most Of Rush ?
```

```python
# Heatmap By Weekday And Hour
```

```python
HeatMap('Day','Hour')
plt.title("Cross Analysis By Hour And Day")
```

```python
#  Each Day In Evening Time There is much Rush
```
